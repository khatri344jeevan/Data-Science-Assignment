```python
# without using libraries

def create_stack():
    stack = []
    return stack

def check_empty(stack):
    return len(stack) == 0

def push(stack,item):
    stack.append(item)
    print("pushed",item)

def pop(stack):
    if (check_empty(stack)):
        return "stack is empty"
    return stack.pop()
stack = create_stack()
push(stack,1)
push(stack,2)
push(stack,3)
push(stack,4)
print("popped",pop(stack))
print("stack",stack)
```

```python
from collections import deque
stack = deque()
stack.append("a")
stack.append("b")
stack.append("c")
stack.append("r")
stack.append("e")

print("Elements popped")
print(stack.pop())
print(stack.pop())
print(stack.pop())

print("Stack:",list(stack))
```
