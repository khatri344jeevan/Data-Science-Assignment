```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

X = np.array([5, 7, 9, 10, 12, 15, 18, 20, 22, 25])
Y = np.array([6, 8, 10, 11, 13, 16, 17, 21, 23, 26])

# Statistics
print("Mean of X:", np.mean(X))
print("Median of X:", np.median(X))
print("Variance of X:", np.var(X))
print("Std Dev of X:", np.std(X))
print("Covariance:", np.cov(X, Y)[0,1])
print("Correlation:", np.corrcoef(X, Y)[0,1])
```

```python
# Histogram
plt.figure()
plt.hist(X, bins=5)
plt.xlabel("Values of X")
plt.ylabel("Frequency")
plt.title("Histogram of X")

plt.show()
```

```python
# Box plot
plt.boxplot(X)
plt.title("Box Plot of X")
plt.show()
```

```python
# Scatter plot
plt.scatter(X, Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot of X vs Y")
plt.show() 
```

```python
# Correlation matrix
df = pd.DataFrame({"X": X, "Y": Y})
print(df.corr())
```
